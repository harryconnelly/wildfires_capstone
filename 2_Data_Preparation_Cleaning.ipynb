#### Round data to nearest 5, I want usa to be categorized like a checker board by 5 latitude by 5 longitude squares. Latitude is ~70 miles and Longitude is ~ 55 miles. Then concat new values with data.

```python
df.head()
```

```python
#chose to append rather than df2['lat'][i] = i because of how much faster it is.
latty5 = []
latty2 = []
latty1 = []
longy5 = []
longy2 = []
longy1 = []

for y, i in enumerate(df['LATITUDE']):
    if y%2500==0:
        print(y)
    latty5.append(int(5*round(float(i)/5)))
    latty2.append(int(2*round(float(i)/2)))
    latty1.append(int(1*round(float(i)/1)))
    longy1.append(int(1*round(float(df['LONGITUDE'][y])/1)))
    longy2.append(int(2*round(float(df['LONGITUDE'][y])/2)))
    longy5.append(int(5*round(float(df['LONGITUDE'][y])/5)))
```

```python
#convert rounded lattitudes to dataframe
df5=pd.DataFrame(latty5).rename(columns={0: 'latt5'})
df6 = pd.DataFrame(latty2).rename(columns={0: 'latt2'})
df7= pd.DataFrame(latty1).rename(columns={0: 'latt1'})
df8=pd.DataFrame(longy1).rename(columns={0: 'longg1'})
df9=pd.DataFrame(longy2).rename(columns={0: 'longg2'})
df11=pd.DataFrame(longy5).rename(columns={0: 'longg5'})
```

```python
#changed over to df10, wanted to keep the df and df2 just incase something needs to be changed I did not want to overwrite it
df10 = pd.concat([df5, df], axis=1)
df10=pd.concat([df6, df10], axis=1)
df10=pd.concat([df7, df10], axis=1)
df10=pd.concat([df8, df10], axis=1)
df10=pd.concat([df9, df10], axis=1)
df10=pd.concat([df11, df10], axis=1)
```
