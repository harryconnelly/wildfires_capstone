```python
import pandas as pd
```

```python
df = pd.read_csv('lagging.csv')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
df2 = pd.read_csv('df2_bigtime.csv')
```

```python
import seaborn as sns
```

#### Model with sns: 

```python
#for size. A, B, C... G
sns.countplot(df2['FIRE_SIZE_CLASS'], order=['A', 'B', 'C', 'D', 'E', 'F', 'G'])
sns.set(font_scale=1.4)
```

```python
#Year distribution
sns.countplot(df['year'])
sns.set(font_scale=1.55)
```

```python
#day of week by sum boxplot

fig, ax = plt.subplots()
sns.boxplot(x="day of week",y='FIRE SIZE SUM', data=df, order = ['Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']).set(xlabel='DAY OF WEEK', ylabel='FIRE  SIZE  SUM  ACRES')
sns.set(font_scale=1.2)
fig.set_size_inches(9, 5)
```

```python
## latitude distribution

fig, ax = plt.subplots()
sns.countplot(df['latit'])
sns.set(font_scale=2)
fig.set_size_inches(9, 5)
```

```python
##month distribution

fig, ax = plt.subplots()
sns.boxplot(df['month'], df['FIRE SIZE MAX']).set(ylabel='FIRE SIZE MAX ACRES')
sns.set(font_scale=1.2)
fig.set_size_inches(9, 5)
```

```python
##day of week

fig, ax = plt.subplots()
sns.countplot(df['day of week'], order = ['Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'])
sns.set(font_scale=1.2)
fig.set_size_inches(9, 5)
```

```python
#year distribution
sns.boxplot(x="year",y='FIRE SIZE MAX', data=df)
```

```python
##fire size sum distribution

sns.barplot(x="year",y='FIRE SIZE SUM', data=df)
```
